```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
import xgboost as xgb
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('train.csv')
```

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rf = RandomForestClassifier(n_estimators = 200, max_depth = 50)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

**XGBoost**

```python
XGB = xgb.XGBClassifier(objective="binary:logistic", use_label_encoder=False)
XGB.fit(X_train, y_train)
```

```python
y_pred = XGB.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

**ANN**

```python
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = keras.Sequential([
    keras.layers.Dense(16, activation='relu', input_shape=(X_train.shape[1],)),  # Hidden layer 1
    keras.layers.Dense(8, activation='relu'),  # Hidden layer 2
    keras.layers.Dense(1, activation='sigmoid')  # Output layer (Sigmoid for binary classification)
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=4, verbose=1)
```

```python
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```
